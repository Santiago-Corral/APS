# Introducción 

En esta primera entrega se diseñó un generador de señales. Se diseñaron dos funciónes que permiten generar señales senodiales y cuadradas, permitiendo ingresar como los siguientes parametros:

La amplitud máxima de la senoidal (volts).
Su valor medio (volts).
La frecuencia (Hz).
La fase (radianes).
La cantidad de muestras digitalizada por el ADC (# muestras).
La frecuencia de muestreo del ADC.


## Función de generación de señales senoidales

```python
def mi_funcion_sen (vmax, dc, ff, ph, nn, fs):
    ts = 1/fs # tiempo de muestreo
    
    # grilla de sampleo temporal
    tt = np.linspace(0, (N-1)*ts, N) #Grilla total de tiempo 
    
    xx = vmax*np.sin(2*np.pi*ff*tt+ph) + dc #Sen (W0 . t + fase) // W0 = 2pi . t 
    
    return tt, xx 
```

## Función de generación de señales cuadradas

```python
def mi_funcion_square (vmax, dc, ff, ph, nn, fs):
    ts = 1/fs # tiempo de muestreo
    
    # grilla de sampleo temporal
    tt = np.linspace(0, (N-1)*ts, N) #Grilla total de tiempo 
    
    xx = vmax * np.sign(np.sin(2 * np.pi * ff * tt + ph)) + dc #Existen funciones que generan squares, pero asi es mas facil
    
    return tt, xx
```
