## Introducción

#### Estimación espectral

La estimación espectral consiste en aproximar la respuesta en frecuencia de un sistema a partir de un conjunto de muestras. En la práctica, la estimación espectral se realiza a partir de una señal discreta, muestreada, y por ende esta sujeta a errores. El periodograma es la técnica más básica  para estimar la densidad espectral de potencia, consiste en calcular la transformada de Fourier discreta de la señal (DFT) y evaluar el cuadrado de su magnitud. 

$$P_{xx} = |{X(f)}|^2$$

Esta tecnica presenta una alta varianza y sufre de leakage espectral, dado que la señal analizada generalmente no es periodica y se producen discontinuidades en los bordes de nuestra ventana de observación. Y una resolución espectral la cual depende directamente de la longitud total de la señal analizada. 

El método de Welch en cambio mejora el periodograma a cambio de una perdida de resolución. En este se divide la señal en segmentos mas cortos con solapamiento entre ellos, luego se ventanea cada uno y se promedian los periodogramas. De esta manera se reduce la varianza del estimador gracias al promedio y mejoramos el efecto de leakage gracias a la implementación de ventanas. 

$$ \hat{S}_{xx}^{\text{Welch}}(f) = \frac{1}{K} \sum_{k=1}^{K} \hat{P}_{xx}^{(k)} (e^{j\omega}) * Window$$

Siendo $ \hat{P}_{xx}^{(k)}$ el periodograma de cada segmento

Por otro lado, antes de aplicar este metodo se corroboro que efectivamente la energía de la señal en el tiempo sea igual a la energía en el dominio de la frecuencia. La identidad de parseval establece una conexión directa entre ambas, lo cual nos indica que la estimación espectral es consistente y adecuada. 

Identidad de parseval:

$$ \int_{-\infty}^{\infty} |x(t)|^2 \, dt = \frac{1}{2\pi} \int_{-\infty}^{\infty} |X(j\omega)|^2 \, d\omega $$

## Código y Análisis

Para el análisis espectral de las señales se decidió estimar con el método de Welch y con la ventana Hamming. Esto nos da una relación varianza, resolución espectral la cual se considero adecuada dados los objetivos del análisis. La estimación por Welch mejora en gran medida la varianza gracias a el promedio de segmentos. Además la ventana reduce los efectos del solapamiento y leakage espectral de los mismos. Se varió el solapamiento de los segmentos para cada señal según se consideraba optima la representación del espectro. 
- Señal de ECG: N/6
- Señal de PPG: N/32
- Señal de Audio: N/16
  
(Siendo N el numero total de muestras)

Para cada señal se verificó el Teorema de Parseval, dado que se analiza el ancho de banda según la potencia acumulada en frecuencia, se procura de esta manera analizar la correcta densidad espectral de potencia, verificando su relación con la señal temporal. 



Importo librerías:

```python
import scipy as sp
import numpy as np

import matplotlib.pyplot as plt
   
import scipy.io as sio

def vertical_flaten(a):
    return a.reshape(a.shape[0],1)
```

#### ECG
Lectura de datos:

```python
fs_ecg = 1000 # Hz, ECG

#ECG Con ruido
mat_struct = sio.loadmat('./ECG_TP4.mat')
ecg_one_lead = vertical_flaten(mat_struct['ecg_lead'])
N1 = len(ecg_one_lead)
ecg_one_lead_n = ecg_one_lead/np.std(ecg_one_lead) #normalizo, divido por el maximo de la señal

plt.figure()
plt.plot(ecg_one_lead_n)
plt.ylabel('Amplitud [mV]')
plt.xlabel('Tiempo [s]')
plt.title("Electrocardiograma (ECG) - Con Ruido")
plt.grid()

#ECG sin ruido
ecg_one_lead = np.load('ecg_sin_ruido.npy')
N2 = len(ecg_one_lead)
ecg_one_lead = ecg_one_lead/np.std(ecg_one_lead) #normalizo, divido por el maximo de la señal

plt.figure()
plt.plot(ecg_one_lead)
plt.ylabel('Amplitud [mV]')
plt.xlabel('Tiempo [s]')
plt.title("Electrocardiograma (ECG) - Sin Ruido");
plt.grid()
```

Estimo la respuesta en frecuencia con el método de Welch:

```python
(fwelch_ecgn, pxx_ecgn) = sp.signal.welch(ecg_one_lead_n, fs_ecg ,nfft = N1, window = 'hamming', nperseg = N1//6, axis = 0)
(fwelch_ecg, pxx_ecg) = sp.signal.welch(ecg_one_lead, fs_ecg ,nfft = N2, window = 'hamming', nperseg = N2//6, axis = 0)

#Ecg con ruido
plt.figure()
bfrec = fwelch_ecgn <= fs_ecg/2
plt.plot(fwelch_ecgn, 10*np.log10(2*np.abs(pxx_ecgn[bfrec])**2))
plt.ylabel('Densidad de Potencia [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.title("Representación espectral - ECG Con Ruido")

#Ecg sin ruido
plt.figure()
bfrec = fwelch_ecg <= fs_ecg/2  
plt.plot(fwelch_ecg, 10*np.log10(2*np.abs(pxx_ecg[bfrec])**2))
plt.ylabel('Densidad de Potencia [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.title("Representación espectral - ECG Sin Ruido");
```

Verifico el teorema de Parseval:

```python
#ECG con ruido
area = np.sum(ecg_one_lead_n**2)

fftecg = np.fft.fft(ecg_one_lead_n.flatten())
energia_espectral = np.mean(np.abs(fftecg)**2) #Mas sencillamente con el valor medio (ver notas)

parseval_ecg_n = np.abs(area - energia_espectral)
print(parseval_ecg_n)

#ECG sin ruido
area = np.sum(ecg_one_lead**2)

fftecg = np.fft.fft(ecg_one_lead.flatten())
energia_espectral = np.mean(np.abs(fftecg)**2) #Mas sencillamente con el valor medio (ver notas)

parseval_ecg = np.abs(area - energia_espectral)
print(parseval_ecg)
```

Buscamos un error cercano a cero, lo cual verifica nuestros resultados. 

Calculamos el ancho de banda:


Buscamos el índice de la sumatoria en el que se alcanzan valores del 95% y 98% de la potencia acumulada total, entendiendo que la potencia e información de la señal real se encuentra comprendido en ese rango de frecuencias. 

```python
#Potencia total de la señal 
potencia_total_ecgn = np.sum(pxx_ecgn) #Devuelve un valor
potencia_total_ecg = np.sum(pxx_ecg)
    
#Potencia acumulada
potencia_acumulada_ecgn = np.cumsum(pxx_ecgn) #Devuelve un vector con los pasos de la sumatoria
potencia_acumulada_ecg = np.cumsum(pxx_ecg)

#Umbrales    
umbral_95_n = 0.95 * potencia_total_ecgn
umbral_98_n = 0.98 * potencia_total_ecgn
    
umbral_95 = 0.95 * potencia_total_ecg
umbral_98 = 0.98 * potencia_total_ecg

#Indice en el que se llega al umbral
idc_95_n = np.argmax(potencia_acumulada_ecgn >= umbral_95_n)
idc_98_n = np.argmax(potencia_acumulada_ecgn >= umbral_98_n)

idc_95 = np.argmax(potencia_acumulada_ecg >= umbral_95)
idc_98 = np.argmax(potencia_acumulada_ecg >= umbral_98)
    
#Bw
BW_95_n = fwelch_ecgn[idc_95_n] #Freq a la que se llega al indice
BW_98_n = fwelch_ecgn[idc_98_n]

BW_95 = fwelch_ecg[idc_95]
BW_98 = fwelch_ecg[idc_98]

#Grafico BW
#ECG con ruido
plt.figure()
bfrec = fwelch_ecgn <= fs_ecg/2
plt.plot(fwelch_ecgn, 10*np.log10(2*np.abs(pxx_ecgn[bfrec])**2))
plt.ylabel('Densidad de Potencia [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.title("Representación espectral - ECG Con Ruido")
plt.xlabel('frequency [Hz]')
plt.axvline(BW_95_n, color='r', linestyle='--', label=f'BW 95%: {BW_95_n:.1f} Hz')
plt.axvline(BW_98_n, color='g', linestyle='--', label=f'BW 98%: {BW_98_n:.1f} Hz')
plt.legend()

#ECG sin ruido
plt.figure()
bfrec = fwelch_ecg <= fs_ecg/2
plt.plot(fwelch_ecg, 10*np.log10(2*np.abs(pxx_ecg[bfrec])**2))
plt.ylabel('Densidad de Potencia [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.title("Representación espectral - ECG Sin Ruido")
plt.xlabel('frequency [Hz]')
plt.axvline(BW_95, color='r', linestyle='--', label=f'BW 95%: {BW_95:.1f} Hz')
plt.axvline(BW_98, color='g', linestyle='--', label=f'BW 98%: {BW_98:.1f} Hz')
plt.legend();
```

```python
#Potencia total de la señal 
potencia_total_c = np.sum(pxx_c) #Devuelve un valor
potencia_total_p = np.sum(pxx_p) 
potencia_total_s = np.sum(pxx_s) 
    
#Potencia acumulada
potencia_acumulada_c = np.cumsum(pxx_c) #Devuelve un vector con los pasos de la sumatoria
potencia_acumulada_p = np.cumsum(pxx_p) 
potencia_acumulada_s = np.cumsum(pxx_s)

#Umbrales    
umbral_95_c = 0.95 * potencia_total_c
umbral_98_c = 0.98 * potencia_total_c
    
umbral_95_p = 0.95 * potencia_total_p
umbral_98_p = 0.98 * potencia_total_p
    
umbral_95_s = 0.95 * potencia_total_s
umbral_98_s = 0.98 * potencia_total_s

#Indice en el que se llega al umbral
idc_95_c = np.argmax(potencia_acumulada_c >= umbral_95_c)
idc_98_c = np.argmax(potencia_acumulada_c >= umbral_98_c)

idc_95_p = np.argmax(potencia_acumulada_p >= umbral_95_p)
idc_98_p = np.argmax(potencia_acumulada_p >= umbral_98_p)

idc_95_s = np.argmax(potencia_acumulada_s >= umbral_95_s)
idc_98_s = np.argmax(potencia_acumulada_s >= umbral_98_s)
    
#Bw
BW_95_c = fwelch_c[idc_95_c] #Freq a la que se llega al indice
BW_98_c = fwelch_c[idc_98_c]

BW_95_p = fwelch_p[idc_95_p]
BW_98_p = fwelch_p[idc_98_p]

BW_95_s = fwelch_s[idc_95_s]
BW_98_s = fwelch_s[idc_98_s]

#Grafico BW
#Audio C
plt.figure()
bfrec = fwelch_c <= fs_audio_c/2
plt.plot(fwelch_c, 10*np.log10(2*np.abs(pxx_c[bfrec])**2))
plt.ylabel('Densidad de Potencia [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.title("Representación espectral - Audio C")
plt.xlabel('frequency [Hz]')
plt.axvline(BW_95_c, color='r', linestyle='--', label=f'BW 95%: {BW_95_c:.1f} Hz')
plt.axvline(BW_98_c, color='g', linestyle='--', label=f'BW 98%: {BW_98_c:.1f} Hz')
plt.legend()

#Audio P
plt.figure()
bfrec = fwelch_p <= fs_audio_p/2
plt.plot(fwelch_p, 10*np.log10(2*np.abs(pxx_p[bfrec])**2))
plt.ylabel('Densidad de Potencia [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.title("Representación espectral - Audio P")
plt.xlabel('frequency [Hz]')
plt.axvline(BW_95_p, color='r', linestyle='--', label=f'BW 95%: {BW_95_p:.1f} Hz')
plt.axvline(BW_98_p, color='g', linestyle='--', label=f'BW 98%: {BW_98_p:.1f} Hz')
plt.legend()

#Audio S
plt.figure()
bfrec = fwelch_s <= fs_audio_s/2
plt.plot(fwelch_s, 10*np.log10(2*np.abs(pxx_s[bfrec])**2))
plt.ylabel('Densidad de Potencia [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.title("Representación espectral - Audio S")
plt.xlabel('frequency [Hz]')
plt.axvline(BW_95_s, color='r', linestyle='--', label=f'BW 95%: {BW_95_s:.1f} Hz')
plt.axvline(BW_98_s, color='g', linestyle='--', label=f'BW 98%: {BW_98_s:.1f} Hz')
plt.legend();
```

## Conclusiones

En primer lugar, experimentalmente se pudo observar una gran dependencia de los valores estimados con la ventana elegida en nuestro estimador y la longitud de los segmentos a promediar, además del solapamiento entre los mismos. Si no se modifica la frecuencia de muestreo, vemos una perdida notoria de resolución espectral, resultando en un espectro menos limpio, irregular y con perdidas ya que perdemos información y discriminación en frecuencia. Además, nuestro método resulta realemente sensible a la ventana elegida, por lo que tenemos que procurar elegir con criterio una ventana con un ancho del lóbulo central adecuado según la resolución que busquemos. Teniendo en cuenta el posible ruido de nuestra señal y la supresión de los lóbulos secundarios que nos puede proveer la ventana, reduciendo el impacto del mismo. Elegir correctamente este balance entre resolución y atenuación en los lóbulos secundarios resulta esencial a la hora de estimar correctamente. 

#### ECG

En principio, podemos observar una importante diferencia en amplitud en la señal con ruido en comparación con la señal limpia de aproximadamente 2mV. Si analizamos el dominio temporal en un mismo intervalo de tiempo podemos ver con claridad oscilaciones de alta frecuencia, las cuales pueden llegar a entorpecer la lectura de la onda "P". Además vemos un desplazamiento (offset) correspondiente a una componente de continua en la señal.

A la hora de estimar el ancho de banda vemos que el ancho de banda del ECG sin ruido (BW95% ≈ 23.4 Hz) es mayor que la señal ruidosa (BW95% ≈ 17.4 Hz). Esto es atribuible a que el ruido esta distriubuido ampliamente en el espectro, esto causa que la potencia de ruido se mezcle con la potencia de la señal real, llegando al umbral buscado de potencia mas rapidamente. 

#### PPG

Al analizar la pletismografía podemos ver que la densidad espectral de potencia se concentra en el rango entre 0 y 10 Hz, lo cual es un valor esperable teniendo en cuenta las frecuencias predominantes de la señal. Al analizar un periodo corto de tiempo se puede detectar e identificar correctamente la forma de la onda PPG, pero con altas componentes de ruido de alta frecuencia ya que vemos la señal con pequeñas oscilaciones erraticas, o "peluda". Estas oscilaciones se pueden ver también al analizar el espectro en frecuencia, ya que vemos picos muy marcados en los rangos de frecuencias entre 200 y 300 Hz, y en 500 Hz. Lectura de datos:

```python
fs_ret = 1700 # Hz

#%% Adquisicion y ordenamiento de datos (GPT)
file_path = r"C:\Users\santi\Desktop\Facu\APS\Trabajos\0001.csv"

# Cargar los datos ignorando la primera línea (encabezado)
datos = np.loadtxt(file_path, delimiter=',', skiprows=1, dtype=str)

# Extraer fecha y hora
fechas_horas = datos[:, 0]
# Extraer los datos (convertidos a float)
re_1 = datos[:, 1].astype(float)
le_1 = datos[:, 2].astype(float)

# Si necesitas ordenarlos por fecha-hora (aunque ya vienen ordenados en este archivo)
# Primero convertir a datetime si es necesario, por ejemplo con datetime de Python

from datetime import datetime

# Convertimos las cadenas de fecha-hora a objetos datetime
fechas_horas_dt = [datetime.strptime(f, "%Y-%m-%d %H:%M:%S.%f") for f in fechas_horas]

# Ordenamos por fecha-hora
indices_ordenados = np.argsort(fechas_horas_dt)
fechas_horas_ordenadas = [fechas_horas_dt[i] for i in indices_ordenados]
re_1_ordenado = re_1[indices_ordenados] 
le_1_ordenado = le_1[indices_ordenados]

#%%

fs_re_1_ordenado = 1700 
N = 255 #Ctd de muestras 255
f0 = fs_re_1_ordenado / N

#Respuesta ojo derecho (re_1_ordenado)

grilla_temporal = np.linspace(0, 1/f0, N)

#Normalizo
re_1_ordenado = re_1_ordenado / np.max(re_1_ordenado)

plt.figure()
plt.plot(grilla_temporal,re_1_ordenado)
plt.ylabel('Amplitud [mV]')
plt.xlabel('Tiempo [s]')
plt.title("Electroretinograma (Ojo derecho)");
plt.grid()
```

Estimo la respuesta en frecuencia con el método de Welch:

```python
N = len(re_1_ordenado)
df = fs_re_1_ordenado/N
(fwelch_re_1_ordenado , pxx_re_1_ordenado) = sp.signal.welch(re_1_ordenado, fs_re_1_ordenado , nfft = N, window = 'hamming', nperseg = N//32, axis = 0)

#re_1_ordenado Espectro

plt.figure()
bfrec = fwelch_re_1_ordenado <= fs_re_1_ordenado/2
plt.plot(fwelch_re_1_ordenado, 10*np.log10(2*np.abs(pxx_re_1_ordenado[bfrec])**2))
plt.ylabel('Densidad de Potencia [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.title("Representación espectral - Electroretinograma (Ojo derecho)");
plt.grid()
```

Verifico el teorema de Parseval:

```python
#re_1_ordenado sin ruido
area = np.sum(re_1_ordenado**2)

fftre_1_ordenado = np.fft.fft(re_1_ordenado.flatten())
energia_espectral = np.mean(np.abs(fftre_1_ordenado)**2) #Mas sencillamente con el valor medio (ver notas)

parseval_re_1_ordenado = np.abs(area - energia_espectral)

print (parseval_re_1_ordenado)
```

Buscamos un error cercano a cero, lo cual verifica nuestros resultados. 



Calculamos el ancho de banda:


Buscamos el índice de la sumatoria en el que se alcanzan valores del 95% y 98% de la potencia acumulada total, entendiendo que la potencia e información de la señal real se encuentra comprendido en ese rango de frecuencias. 

```python
#%% Calculo ancho de banda
#Sumamos potencia acumulada hasta llegar a un 95% 
#(consideramos este punto como la pot. total del sistema)

#Potencia total de la señal 
potencia_total_re_1_ordenado = np.sum(pxx_re_1_ordenado) #Devuelve un valor
    
#Potencia acumulada
potencia_acumulada_re_1_ordenado = np.cumsum(pxx_re_1_ordenado) #Devuelve un vector con los pasos de la sumatoria

#Umbrales    
umbral_95 = 0.95 * potencia_total_re_1_ordenado
umbral_98 = 0.98 * potencia_total_re_1_ordenado

#Indice en el que se llega al umbral
idc_95 = np.argmax(potencia_acumulada_re_1_ordenado >= umbral_95)
idc_98 = np.argmax(potencia_acumulada_re_1_ordenado >= umbral_98)
    
#Bw
BW_95 = fwelch_re_1_ordenado[idc_95]
BW_98 = fwelch_re_1_ordenado[idc_98]

#Grafico BW
#ECG con ruido
plt.figure()
bfrec = fwelch_re_1_ordenado <= fs_re_1_ordenado/2
plt.plot(fwelch_re_1_ordenado, 10*np.log10(2*np.abs(pxx_re_1_ordenado[bfrec])**2))
plt.ylabel('Densidad de Potencia [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.title("Representación espectral - Electroretinograma (Ojo derecho)")
plt.xlabel('frequency [Hz]')
plt.axvline(BW_95, color='r', linestyle='--', label=f'BW 95%: {BW_95:.1f} Hz')
plt.axvline(BW_98, color='g', linestyle='--', label=f'BW 98%: {BW_98:.1f} Hz')
plt.legend();
plt.grid()
```
