## Introducción

#### Transformada $\mathcal{Z}$

La transformada $\mathcal{Z}$ corresponde a evaluar una función en el plano S (análogo discreto de la transformada de Laplace) y para Fourier el eje imaginario $s = j \omega$ se transforma en el círculo unitario en el plano Z. Por lo que la transformada permite representar señales discretas usando la variable compleja:

$$ \mathcal{Z} = re^{j\omega} $$

Con r = modulo del radio variable. Y donde la transformada $\mathcal{Z}$ para sistemas discretos resulta:

$$ \mathcal{Z}\{x(n)\} = X(z) = \sum_{n=0}^{\infty} x(n) z^{-n} $$

La transformada exige que la serie de la suma converga, por lo que existe una región de convergencia (ROC). Un sistema es estable si su ROC incluye al circulo unitario. Por lo que al evaluar la transformada sobre el eje $\mathcal{Z} = re^{j\omega$ la serie converge. Al trabajar con funciones transferencia de la forma: 

$$
H(z) = \frac{Y(z)}{X(z)} = \frac{b_0 + b_1 z^{-1} + b_2 z^{-2} + \cdots + b_M z^{-M}}{1 + a_1 z^{-1} + a_2 z^{-2} + \cdots + a_N z^{-N}}
$$

Donde $( b_i )$ y $( a_i )$ determinan los ceros y los polos del sistema respectivamente y $z^{-k} $ representa un retardo de $( k )$ muestras. Teniendo en cuenta la ROC veamos donde se encuentran los polos y los ceros con respecto al radio unitario: 

- Si todos los polos están dentro del círculo unitario (|z| < 1), el sistema es estable.
- Si uno o más polos están fuera del círculo unitario (|z| > 1), el sistema es inestable.
- Si alguno está sobre el círculo (|z| = 1), el sistema es marginalmente estable.

#### Transformada y respuesta en frecuencia de modulo y fase

Luego de transformar y obtener la función transferencia, evaluamos la función sobre el radio unitario $z = e^{j\omega}$. De esta manera podemos conocer como se modifica el modulo y la fase de cada componente frecuencial $\omega$ de entrada.

#### a)  $y(n)=x(n−3)+x(n−2)+x(n−1)+x(n)$


1. Transformada $\mathcal{Z}\{y(n)\} $


    $$Y(z) = x(n) * z^{-3} + x(n) * z^{-2} + x(n) * z^{-1} +1$$

3. Función Transferencia

    $$H(z) = \frac{Y(z)}{X(z)} = z^{-3} + z^{-2} + z^{-1} + 1$$

    $$H(z) = \frac { z^{3} + z^{2} + z^{1} + 1 } {z^{3}} $$

4.  Respuesta en frecuencia

    $$ H(e^{j\omega}) = e^{-3j\omega} (1+ e^{j\omega} + e^{2j\omega} + e^{3j\omega}) $$

   $$ 
   H(e^{j\omega}) = e^{-\frac{3}{2}j\omega} (e^{-\frac{3}{2}j\omega} + e^{-\frac{1}{2}j\omega} + e^{\frac{1}{2}j\omega} + e^{\frac{3}{2}j\omega})
   $$

   $$ H(e^{j\omega}) = e^{-\frac{3}{2}j\omega} (2 \; cos({\frac{3}{2}\omega}) +2 \; cos({\frac{1}{2}\omega})) $$   

   Modulo: $$|H(e^{j\omega})| =  (2 \; cos({\frac{3}{2}\omega}) +2 \; cos({\frac{1}{2}\omega})) $$
   Fase: $$H(e^{j\omega}) =  -\frac{3}{2} \omega $$

#### b)  $y(n)=x(n−4)+x(n−3)+x(n−2)+x(n−1)+x(n)$

1. Transformada $\mathcal{Z}\{y(n)\} $

    $$Y(z) = x(n) * z^{-4} + x(n) * z^{-3} + x(n) * z^{-2} + x(n) * z^{-1} + x(n) $$

3. Función Transferencia

    $$H(z) = \frac{Y(z)}{X(z)} = z^{-4} + z^{-3} + z^{-2} + z^{-1} + 1$$

    $$H(z) = \frac { z^{4} + z^{3} + z^{2} + z^{1} + 1 } {z^{4}} $$

4.  Respuesta en frecuencia

    $$ H(e^{j\omega}) = e^{-4j\omega} + e^{-3j\omega} + e^{-2j\omega} + e^{-j\omega} + 1 $$

   $$ 
   H(e^{j\omega}) = e^{-2j\omega} \left( e^{-2j\omega} + e^{-j\omega} + 1 + e^{j\omega} + e^{2j\omega} \right)
   $$

   $$ H(e^{j\omega}) = e^{-2j\omega} \left(2 \; \cos(2\omega) + 2 \; \cos(\omega) + 1\right) $$   

   Módulo: 
   $$|H(e^{j\omega})| =  2 \; \cos(2\omega) + 2 \; \cos(\omega) + 1$$

   Fase: 
   $$\angle H(e^{j\omega}) =  -2 \omega$$

#### c)  $y(n)=x(n)−x(n−1)$

1. Transformada $\mathcal{Z}\{y(n)\} $

    $$Y(z) = x(n) * 1 - x(n) * z^{-1}$$

3. Función Transferencia

    $$H(z) = \frac{Y(z)}{X(z)} = 1 - z^{-1}$$

    $$H(z) = \frac{z - 1}{z}$$

4. Respuesta en frecuencia

   $$H(e^{j\omega}) = 1 - e^{-j\omega}$$

   $$ 
   H(e^{j\omega}) = e^{-j\omega/2} \left( e^{j\omega/2} - e^{-j\omega/2} \right)
   $$

   $$ H(e^{j\omega}) = e^{-j\omega/2} \cdot \left(2j \sin(\omega/2)\right) $$

   Módulo:  
   $$|H(e^{j\omega})| = 2 \; \sin(\omega/2)$$

   Fase:  
   $$\angle H(e^{j\omega}) = -\frac{\omega}{2} + \frac{\pi}{2}$$

#### d)  $y(n)=x(n)−x(n−2)$

1. Transformada $\mathcal{Z}\{y(n)\}$

   $$
   Y(z) = x(n) * 1 - x(n) * z^{-2}
   $$

3. Función Transferencia

   $$
   H(z) = \frac{Y(z)}{X(z)} = 1 - z^{-2}
   $$

   $$
   H(z) = \frac{z^{2} - 1}{z^{2}}
   $$

4. Respuesta en frecuencia

   $$
   H(e^{j\omega}) = 1 - e^{-2j\omega}
   $$

   $$
   H(e^{j\omega}) = e^{-j\omega} \left( e^{j\omega} - e^{-j\omega} \right)
   $$

   $$
   H(e^{j\omega}) = e^{-j\omega} \cdot (2j \sin(\omega))
   $$

   Módulo:  
   $$
   |H(e^{j\omega})| = 2 \; \sin(\omega)
   $$

   Fase:  
   $$
   \angle H(e^{j\omega}) = -\omega + \frac{\pi}{2}
   $$

#### Aclaración
Para evaluar la respuesta en frecuencia de modulo y fase se busco sacar factor común $e^{-jomega \frac{N}{2}}$ donde N es el retardo máximo de la fase. Esto resulta de gran ayuda ya que los filtros no recursivos (FIR) tienen simetría en la mitad de las muestras. Lo cual facilita luego las cuentas para la respuesta en frecuencia. También se hizo uso de la identidad de Euler: $$
\cos(\omega) = \frac{e^{j\omega} + e^{-j\omega}}{2} \quad\quad \sin(\omega) = \frac{e^{j\omega} - e^{-j\omega}}{2j}
$$




## Código y Análisis

Importo librerías:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Importo Función para analizar, modulo, fase y diagrama de polos y ceros de mi función transferencia (H)

```python
def Analisis_Frecuencial(H, Numeradores, Denominadores):
    omega = np.linspace(0, np.pi, 1000) #Frecuencias normalizadas

    H_Modulo = np.abs(H)
    H_Fase = np.angle(H)

    # Graficos
    
    plt.figure(figsize=(18, 5))

    # Subplot 1: Módulo
    plt.subplot(1, 3, 1)
    plt.plot(omega, H_Modulo)
    plt.title('Módulo de H(e^{jω})')
    plt.xlabel('ω [rad/muestra]')
    plt.ylabel('|H(e^{jω})|')
    plt.grid(True)

    # Subplot 2: Fase
    plt.subplot(1, 3, 2)
    plt.plot(omega, H_Fase)
    plt.title('Fase de H(e^{jω})')
    plt.xlabel('ω [rad/muestra]')
    plt.ylabel('Fase [rad]')
    plt.grid(True)

    # Subplot 3: Diagrama de polos y ceros
    
    #POLOS y CEROS
    ceros = np.roots(Numeradores)
    polos = np.roots(Denominadores)

    # Círculo unitario
    theta = np.linspace(0, 2*np.pi, 300)
    circ_x = np.cos(theta)
    circ_y = np.sin(theta)
    
    plt.subplot(1, 3, 3)
    plt.plot(circ_x, circ_y, 'k--', label='Círculo unitario')
    plt.plot(np.real(ceros), np.imag(ceros), 'o', label='Ceros', markersize=8, markerfacecolor='none')
    plt.plot(np.real(polos), np.imag(polos), 'x', label='Polos', markersize=8)
    plt.xlabel('Re')
    plt.ylabel('Im')
    plt.title('Polos y Ceros')
    plt.axis('equal')
    plt.grid(True)
    plt.legend(loc='upper left')

    plt.tight_layout()
    plt.show()
```

Evaluo las transferencias

```python
omega = np.linspace(0, np.pi, 1000) #Frecuencias normalizadas

H_a = 1 + np.exp(-1j*omega) + np.exp(-2j*omega) + np.exp(-3j*omega)
Numeradores_a = [1, 1, 1, 1]
Denominadores_a = [1, 0, 0, 0]

H_b = 1 + np.exp(-1j*omega) + np.exp(-2j*omega) + np.exp(-3j*omega) + np.exp(-4j*omega)
Numeradores_b = [1, 1, 1, 1, 1]
Denominadores_b = [1, 0, 0, 0, 0]

H_c = 1 - np.exp(-1j*omega)
Numeradores_c = [1, -1]
Denominadores_c = [1, 0]

H_d = 1 - np.exp(-2j*omega)
Numeradores_d = [1, 0 , -1]
Denominadores_d = [1, 0, 0]
```

a)

```python
Analisis_Frecuencial(H_a, Numeradores_a, Denominadores_a)
```

b)

```python
Analisis_Frecuencial(H_b, Numeradores_b, Denominadores_b)
```

c)

```python
Analisis_Frecuencial(H_c, Numeradores_c, Denominadores_c)
```

d)

```python
Analisis_Frecuencial(H_d, Numeradores_d, Denominadores_d)
```

## Conclusiones

Podemos entender el comportamiento de cada uno de nuestros sistemas a partir de la ubicación de los polos y los ceros del mismo. En cuanto a los ceros, estos actúan como anuladores de ciertas frecuencias, anulando la frecuencia correspondiente al ángulo formado desde el origen hasta la ubicación del cero. Luego los polos nos determinan el orden y estabilidad del sistema, en nuestro caso se trata de sistemas estables dado que los polos se encuentran en el origen.

Entendemos a los filtros FIR (Finit Impulse Response filter) como filtros los cuales no varían dependiendo de salidas pasadas (como un impulso). La salidad depende únicamente de la señal de entrada y sus retardos. Esto es facilmente verificable viendo que las antitransformadas de las expresiónes regresan exactamente a la expresión original, sin terminos de realimentación como si ocurre en  los filtros IIR (Infinit Impulse Response filter). Esto implica un orden mas alto (mas costosos computacionalmente) y menor sensibilidad que los IIR, pero resultan mas estables dado que cuentan con fase lineal. 

#### Filtro a)

En este caso, evaluando la respuesta de modulo, podemos ver que se trata de un filtro pasa-bajos. Con ceros ubicados en $\pi$ y $\pi/2$ por lo que para estos angulos podemos ver como el modulo de la respuesta en frecuencia se anula debido a la interferencia destructiva de los ceros. Los saltos que vemos en la fase también son atribuibles a estas anulaciónes en el módulo. También encontramos una relación entre la pendiente y la cantidad de ceros o proximidad de los ceros al origen. 

#### Filtro b)

En este caso, también se trata de un filtro pasa-bajos, pero con mayor complejidad que el anterior. Tenemos 4 ceros ubicados en angulos intermedios entre $0$ y $\pi$. Esto se traduce en una pendiente mas pronunciada del filtro y mas saltos y transiciones bruscas en la fase. 

#### Filtro c)

Para el filtro c, se corresponde a un filtro pasa altos, este se identifica por tener un unico cero ubicado en el radio unitario. Por lo que las frecuencias bajas son atenuadas y permitiendo el paso de las altas. Al no tener regiones criticas en el radio unitario podemos ver una respuesta de fase lineal y decreciente. 

#### Filtro d)

Por ultiumo este filtro actua como un pasa banda, al tener dos ceros se suprimen frecuencias tanto bajas como altas. Llegando a su maxima amplitud en $\pi/2$. Podemos ver que también se trata de una fase decreciente y lineal, ya que no tenemos interferencias destructivas en los ceros que anulen la respuesta en modulo.
